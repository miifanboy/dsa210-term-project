```python
import pandas as pd
from utils.transformData import json_to_csv

df = pd.read_csv("data/steamlinks.csv")
for row in df.itertuples(index=True, name='Pandas'):
    name = row[1]
    json_to_csv(f"data/raw/{name}.json",f"data/processed1/{name}.csv")
```
