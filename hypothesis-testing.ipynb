# Hypothesis Testing

## These are some Null Hypotheses that I am going to test:
| Question? | Null Hypothesis |
| --------- | --------------- |
| Did Gold prices significantly drop post-update? | There is no difference in mean Gold prices before and after Oct 22, 2025. |
| Did the update change the correlation between rarity classes? | The correlation between Covert and Gold prices is the same before and after the update. |
| Is the Gold market now more volatile? | The variance in Gold prices is unchanged. |

### 1. There is no difference in mean Gold prices before and after Oct 22, 2025.
##### Method of testing: t_test without equal variance

```python
from scipy import stats
import pandas as pd

gold_path = "data/processed1/Falchion Knife - Lore (Field-Tested).csv"
g_df = pd.read_csv(gold_path)

before = g_df[g_df['date'] < '2025-10-22']['price']
after = g_df[g_df['date'] >= '2025-10-22']['price']
t_stat, p_val = stats.ttest_ind(before, after, equal_var=False)

print(f"T-statistic: {t_stat:.3f}, P-value: {p_val:.3f}")
if p_val < 0.05:
    print("Reject H0: Significant change in mean price.")
```

### 2. The correlation between Covert and Gold prices is the same before and after the update.
##### Method of testing: Spearman's Correlation

```python
gold = pd.read_csv(gold_path)
covert = pd.read_csv("data/processed1/AK-47 - Nightwish (Field-Tested).csv")

merged = pd.merge(gold[['date', 'volume']], covert[['date', 'price']], on='date')

corr, p_val = stats.spearmanr(merged['volume'], merged['price'])

print(f"Spearman Correlation: {corr:.4f}, P-value: {p_val:.4f}")
if(corr < 0.5):
    print(f"Poor correlation.")
```

### 3. The variance in Gold prices is unchanged.
##### Method of testing: Levene's test to compare variances

```python
stat, p_val = stats.levene(before, after)

print(f"Levene Statistic: {stat:.3f}, P-value: {p_val:.3f}")
if p_val < 0.05:
    print("Reject H0: Volatility (variance) changed significantly after the update.")
```

## Verdict
### All three hypothesis are rejected. I already expected the first one to be rejected as it is clear that it made an effect but the second hypothesis shows that the market dynamics actually changed significantly. Third one was a bit suprise too.
